```
import pandas as pd
```

```
df_oceans = pd.read_csv('dados_oceans14.csv', encoding='utf-8')
df_oceans = df_oceans[df_oceans['ano'] == 2020]
df_oceans.set_index('tick', inplace=True)
df_bruto_status = pd.read_csv('dados_brutos_status_invest-2020_06_15.csv', encoding='utf-8', sep=';')
```

```
df_oceans.columns
```

```
df_bruto_status.columns
```

```
COLUNAS_UTILIZADAS = ['TICKER', 'DY', 'P/L', 'P/VP', 'P/ATIVOS', 'MARGEM EBIT', 'MARG. LIQUIDA', 'P/EBIT', 'EV/EBIT', 'DIVIDA LIQUIDA / EBIT', 'PSR', 'LIQ. CORRENTE', 'ROE', 'ROA', 'ROIC', 'CAGR LUCROS 5 ANOS', ' LIQUIDEZ MEDIA DIARIA']
COLUNAS_UTILIZADAS_AJEITADAS = ['tick', 'DY', 'P_L', 'P_VP', 'P_ativos', 'margem_EBIT', 'margem_liq', 'P_EBIT', 'EV_EBIT', 'divida_liq_EBIT', 'PSR', 'liq_corrente', 'ROE', 'ROA', 'ROIC', 'lucros_5anos', 'liq_media_diaria']
```

```
df_status = df_bruto_status[COLUNAS_UTILIZADAS].copy()
df_status.columns = COLUNAS_UTILIZADAS_AJEITADAS
df_status.set_index('tick', inplace=True)
df_status
```

```
df_oceans
```

```
colunas_exclusivas_oceans = list(set(df_oceans.columns) - set(COLUNAS_UTILIZADAS_AJEITADAS))
df_oceans = df_oceans[colunas_exclusivas_oceans]
df_oceans
```

```
df_consolidado = df_oceans.join(df_status).copy()
df_consolidado.to_csv('dados_oceans_status.csv', encoding='utf-8', index=False)
df_consolidado
```
